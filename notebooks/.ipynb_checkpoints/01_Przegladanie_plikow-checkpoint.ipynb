```python
#interpolator = sitk.sitkLinear
interpolator = sitk.sitkNearestNeighbor
#ransform = sitk.ReadTransform(t2_iter_path)

def resample_to_reference(image: sitk.Image, reference: sitk.Image, transform: sitk.Transform = sitk.Transform(),
                          interpolator=sitk.sitkBSpline):
        r = sitk.ResampleImageFilter()
        r.SetReferenceImage(reference)
        r.SetTransform(transform)
        r.SetInterpolator(interpolator)
        return r.Execute(image)
sitk_resampled_image = resample_to_reference(sitk_t1, sitk_aseg)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

from skimage import data
from skimage.filters import threshold_otsu
from skimage.segmentation import clear_border
from skimage.measure import label, regionprops
from skimage.morphology import closing, square
from skimage.color import label2rgb

image = x

# apply threshold
thresh = threshold_otsu(image)
bw = closing(image > thresh, square(4))

# remove artifacts connected to image border
cleared = clear_border(bw)

# label image regions
label_image = label(cleared)
# to make the background transparent, pass the value of `bg_label`,
# and leave `bg_color` as `None` and `kind` as `overlay`
image_label_overlay = label2rgb(label_image, image=image, bg_label=0)

fig, ax = plt.subplots(figsize=(10, 6))
ax.imshow(image_label_overlay)

for region in regionprops(label_image):
    # take regions with large enough areas
    if region.area >= 100:
        # draw rectangle around segmented coins
        minr, minc, maxr, maxc = region.bbox
        rect = mpatches.Rectangle((minc, minr), maxc - minc, maxr - minr,
                                  fill=False, edgecolor='red', linewidth=2)
        ax.add_patch(rect)

ax.set_axis_off()
plt.tight_layout()
plt.show()
```

```python
from IPython.display import display
import ipywidgets as widgets
from ipywidgets import interact, interactive

#przekroj = widgets.IntSlider(min =1, max = 150, value  = 75)
#display(przekroj)
#przekroj = 80
    
def plot_func(przekroj, odciecie, dilation_step):
    x = np.where(  (resampled_image_value > 0) & (resampled_image_value < odciecie), 1, 0)[:,:,przekroj]

    image = x
    selem = disk(dilation_step)
    image = dilation(image, selem)

    # apply threshold
    thresh = threshold_otsu(image)
    bw = closing(image > thresh, square(3))

    # remove artifacts connected to image border
    cleared = clear_border(bw)

    # label image regions
    label_image = label(cleared)
    # to make the background transparent, pass the value of `bg_label`,
    # and leave `bg_color` as `None` and `kind` as `overlay`
    image_label_overlay = label2rgb(label_image, image=image, bg_label=0)

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.imshow(image_label_overlay)

    for region in regionprops(label_image):
        # take regions with large enough areas
        if region.area >= 100:
            # draw rectangle around segmented coins
            minr, minc, maxr, maxc = region.bbox
            rect = mpatches.Rectangle((minc, minr), maxc - minc, maxr - minr,
                                      fill=False, edgecolor='red', linewidth=2)
            ax.add_patch(rect)

    ax.set_axis_off()
    plt.tight_layout()
    plt.show()
    
interact(plot_func, przekroj = widgets.IntSlider(min =1, max = 250, value  = 75), 
         odciecie = widgets.IntSlider(min =1, max = 255, value  = 230),
         dilation_step = widgets.FloatSlider(min = 0.01, max = 6, value  = 1))
```

```python
import matplotlib.pyplot as plt
from skimage import data
from skimage.util import img_as_ubyte
from skimage import io
from skimage.morphology import erosion, dilation, opening, closing, white_tophat
from skimage.morphology import black_tophat, skeletonize, convex_hull_image
from skimage.morphology import disk

def plot_comparison(original, filtered, filter_name):

    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 4), sharex=True,
                                   sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('original')
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(filter_name)
    ax2.axis('off')
    
selem = disk(1.5)
dilated = dilation(x, selem)
plot_comparison(x, dilated, 'dilation')
```

```python
selem = disk(2)
eroded = erosion(x, selem)
plot_comparison(x, eroded, 'erosion')
```
